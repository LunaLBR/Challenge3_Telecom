##Modelo

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.metrics import (classification_report, confusion_matrix, roc_auc_score)

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.pipeline import Pipeline
```

```python
# copia del df original

df_model = df.copy()

# 'account_Charges_Total' a numérico (por si hay strings/espacios)
df_model['account_Charges_Total'] = pd.to_numeric(df_model['account_Charges_Total'], errors='coerce')

# Churn (Yes=1 / No=0)
if df_model['Churn'].dtype == 'object':
    df_model['Churn'] = df_model['Churn'].map({'Yes': 1, 'No': 0})

#Definimos columnas numéricas y categóricas
num_cols = ['customer_tenure', 'account_Charges_Monthly', 'account_Charges_Total', 'customer_SeniorCitizen']
cat_cols = [
    'customer_gender','customer_Dependents',
    'phone_PhoneService','phone_MultipleLines',
    'internet_InternetService','internet_OnlineSecurity','internet_OnlineBackup',
    'internet_DeviceProtection','internet_TechSupport',
    'internet_StreamingTV','internet_StreamingMovies',
    'account_Contract','account_PaperlessBilling','account_PaymentMethod'
]
```

```python
#Separación Train/Test (70/30)
X = df_model[num_cols + cat_cols]
y = df_model['Churn'].astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, stratify=y, random_state=42
)
```

```python
# Preprocesador con normalización (para LogReg y KNN)
#   - Se imputan valores faltantes con la mediana (numéricas)


numeric_scaled = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
#   - Se escalan las variables numéricas a media=0, std=1 (StandardScaler)
categorical_onehot = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])
#   - Las variables categóricas se convierten a dummies (OneHotEncoder)
preprocessor_scaled = ColumnTransformer(
    transformers=[
        ('num', numeric_scaled, num_cols),
        ('cat', categorical_onehot, cat_cols)
    ]
)

```

```python
# Preprocesador sin normalización (para Árbol)
#   - Igual que arriba, pero sin StandardScaler en las numéricas
#   - Los árboles no usan distancias ni gradientes, solo cortes en valores
numeric_passthrough = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median'))
])
preprocessor_tree = ColumnTransformer(
    transformers=[
        ('num', numeric_passthrough, num_cols),
        ('cat', categorical_onehot, cat_cols)
    ]
)
```

####Modelos

```python
# 1) Regresión Logística sensible a la escala → usamos preprocessor_scaled
pipe_logreg = Pipeline(steps=[
    ('prep', preprocessor_scaled),
    ('clf', LogisticRegression(max_iter=1000, solver='lbfgs'))
])
```

```python
# 2) KNN sensible a la escala → usamos preprocessor_scaled
pipe_knn = Pipeline(steps=[
    ('prep', preprocessor_scaled),
    ('clf', KNeighborsClassifier(n_neighbors=10))
])
```

```python
# 3) Árbol de Decisión no sensible a la escala → usamos preprocessor_tree
pipe_tree = Pipeline(steps=[
    ('prep', preprocessor_tree),
    ('clf', DecisionTreeClassifier(random_state=42))
])
```

```python
modelos = {
    "Regresión Logística": pipe_logreg,
    "KNN": pipe_knn,
    "Árbol de Decisión": pipe_tree
}

for nombre, modelo in modelos.items():
    print(f"\n{'='*70}\nModelo: {nombre}\n{'='*70}")
    modelo.fit(X_train, y_train)
    y_pred = modelo.predict(X_test)

    print(classification_report(y_test, y_pred, target_names=['No','Yes']))

    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['No','Yes'], yticklabels=['No','Yes'])
    plt.title(f"Matriz de Confusión - {nombre}")
    plt.xlabel("Predicción")
    plt.ylabel("Real")
    plt.show()
     # Precisión (Accuracy): % total de aciertos
    # Recall: qué % de los clientes que cancelan fueron detectados
    # F1-score: balance entre precisión y recall
```

```python
""" RESULTADOS
1. Regresión Logística

Mejor desempeño global (Accuracy = 80%).

Buen equilibrio entre precisión (0.64) y recall (0.55).

F1 = 0.59, el mejor de los tres modelos.

2. KNN

Accuracy un poco menor (77%).

Precisión moderada (0.59), pero recall bajo (0.48).

Está detectando menos clientes que cancelan (muchos falsos negativos en la matriz de confusión, Esto puede deberse a que k no está optimizado

Necesita ajuste de hiperparámetros (k, distancia, pesos). De lo contrario, está en ligero underfitting. (Cuando el modelo no captura bien las tendencias de los datos, lo que indica que es demasiado simple.)

3. Árbol de Decisión

Accuracy más bajo (73%).

Precisión baja (0.49), pero recall ligeramente mayor que KNN (0.52).

El árbol puede estar sobreajustando el train y perdiendo generalización en test. No se limitó la profundidad (max_depth), lo que puede causar overfitting. (Cuando el modelo aprende demasiado sobre los datos de entrenamiento, perdiendo la capacidad de generalizar a nuevos datos.)

"""
```
